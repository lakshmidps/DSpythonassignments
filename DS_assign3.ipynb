```r
# 1. Function to convert Celsius to Fahrenheit
celsius_to_fahrenheit <- function(celsius) {
  fahrenheit <- (celsius * 9/5) + 32
  return(fahrenheit)
}

# Read Celsius temperature from the user
celsius <- as.numeric(readline(prompt = "Enter temperature in Celsius: "))

# Convert Celsius to Fahrenheit
fahrenheit <- celsius_to_fahrenheit(celsius)

# Print the result
cat("The equivalent temperature in Fahrenheit is:", fahrenheit, "\n")
```

```r
# 2. Function to calculate the area of a circle
calculate_area <- function(radius) {
  area <- pi * radius^2
  return(area)
}

# Function to calculate the perimeter (circumference) of a circle
calculate_perimeter <- function(radius) {
  perimeter <- 2 * pi * radius
  return(perimeter)
}

# Read radius of the circle from the user
radius <- as.numeric(readline(prompt = "Enter the radius of the circle: "))

# Calculate the area and perimeter of the circle
area <- calculate_area(radius)
perimeter <- calculate_perimeter(radius)

# Print the results
cat("The area of the circle is:", area, "\n")
cat("The perimeter (circumference) of the circle is:", perimeter, "\n")
```

```r
# 3. Function to calculate the final amount after adding interest
calculate_final_amount <- function(amount, interest_rate) {
  interest = (amount * interest_rate) / 100
  final_amount = amount + interest
  return(final_amount)
}

# Read the original amount from the user
amount <- as.numeric(readline(prompt = "Enter the original amount: "))

# Read the percentage of interest from the user
interest_rate <- as.numeric(readline(prompt = "Enter the percentage of interest: "))

# Calculate the final amount after adding interest
final_amount <- calculate_final_amount(amount, interest_rate)

# Print the result
cat("The final amount after adding interest is:", final_amount, "\n")
```

```r
# 4. Function to convert meters to centimeters
convert_to_centimeters <- function(meters) {
  centimeters <- meters * 100
  return(centimeters)
}

# Function to convert meters to inches
convert_to_inches <- function(meters) {
  inches <- meters * 39.3701
  return(inches)
}

# Function to convert meters to yards
convert_to_yards <- function(meters) {
  yards <- meters * 1.09361
  return(yards)
}

# Read the distance in meters from the user
meters <- as.numeric(readline(prompt = "Enter distance in meters: "))

# Convert meters to centimeters, inches, and yards
centimeters <- convert_to_centimeters(meters)
inches <- convert_to_inches(meters)
yards <- convert_to_yards(meters)

# Print the results
cat("Distance in centimeters:", centimeters, "cm\n")
cat("Distance in inches:", inches, "in\n")
cat("Distance in yards:", yards, "yd\n")
```

```r
# 5. Function to calculate profit or loss
calculate_profit_or_loss <- function(selling_cost, actual_cost) {
  if (selling_cost > actual_cost) {
    profit <- selling_cost - actual_cost
    return(paste("Profit:", profit))
  } else if (selling_cost < actual_cost) {
    loss <- actual_cost - selling_cost
    return(paste("Loss:", loss))
  } else {
    return("No Profit, No Loss")
  }
}

# Read selling cost from the user
selling_cost <- as.numeric(readline(prompt = "Enter the selling cost of the product: "))

# Read actual cost from the user
actual_cost <- as.numeric(readline(prompt = "Enter the actual cost of the product: "))

# Calculate profit or loss
result <- calculate_profit_or_loss(selling_cost, actual_cost)

# Print the result
cat(result, "\n")
```

```r
# 6. Function to calculate gross salary
calculate_gross_salary <- function(basic_salary) {
  # Calculate DA and HRA
  da <- 0.75 * basic_salary
  hra <- 0.20 * basic_salary

  # Calculate gross salary based on the conditions
  if (basic_salary < 10000) {
    gross_salary <- da + basic_salary
  } else if (basic_salary >= 10000 && basic_salary < 20000) {
    gross_salary <- da + basic_salary + (0.50 * hra)
  } else {
    gross_salary <- basic_salary + da + hra
  }

  return(gross_salary)
}

# Read basic salary from the user
basic_salary <- as.numeric(readline(prompt = "Enter the basic salary: "))

# Calculate gross salary
gross_salary <- calculate_gross_salary(basic_salary)

# Print the result
cat("The gross salary is:", gross_salary, "\n")
```

```r
# 7. Function to calculate factorial
calculate_factorial <- function(number) {
  # Initialize factorial to 1
  factorial <- 1

  # Calculate factorial using a loop
  for (i in 1:number) {
    factorial <- factorial * i
  }

  return(factorial)
}

# Read a number from the user
number <- as.numeric(readline(prompt = "Enter a number: "))

# Validate that the input is a non-negative integer
if (number < 0 || floor(number) != number) {
  cat("Please enter a non-negative integer.\n")
} else {
  # Calculate factorial
  result <- calculate_factorial(number)

  # Print the result
  cat("The factorial of", number, "is:", result, "\n")
}
```

```r
# 8. Function to calculate the sum of all numbers in the range
calculate_sum_in_range <- function(num1, num2) {
  # Ensure num1 is less than or equal to num2
  if (num1 > num2) {
    temp <- num1
    num1 <- num2
    num2 <- temp
  }

  # Calculate the sum of all numbers in the range
  sum <- sum(num1:num2)
  return(sum)
}

# Read num1 and num2 from the user
num1 <- as.numeric(readline(prompt = "Enter the first number (num1): "))
num2 <- as.numeric(readline(prompt = "Enter the second number (num2): "))

# Calculate the sum of all numbers between num1 and num2
result <- calculate_sum_in_range(num1, num2)

# Print the result
cat("The sum of all numbers between", num1, "and", num2, "is:", result, "\n")
```

```r
# Function to check if a number is odd or even
check_odd_even <- function(number) {
  if (number %% 2 == 0) {
    return("even")
  } else {
    return("odd")
  }
}

# Read a number from the user
number <- as.numeric(readline(prompt = "Enter a number: "))

# Check if the number is odd or even using the function
result <- check_odd_even(number)

# Print the result
cat("The number", number, "is", result, "\n")
```
